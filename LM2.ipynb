```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
features_drop = ["PassengerId","Pclass",'Age','Fare','Family']
train_label = train['Survived']
train_features = train.drop(features_drop, axis=1)
test_features = test.drop(features_drop, axis=1)
train_features = train_features.drop("Survived", axis=1)

print(train_features.shape, train_label.shape, test_features.shape)
```

```python
train_features.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.utils import shuffle
```

```python
train_features, train_label = shuffle(train_features, train_label, random_state = 121)
```

```python
def train_test(model):

    model.fit(train_features, train_label)
    prediction = model.predict(test_features)
    accuracy = round(model.score(train_features, train_label) * 100, 2)
    print("Accuracy : ", accuracy, "%")
    return prediction
```

```python
LM_pred = train_test(LogisticRegression(C= 100))

```

```python
submission = pd.DataFrame({"PassengerId" : test["PassengerId"],
                           "Survived" : LM_pred})
submission.to_csv('C:/Users/JH/Desktop/submission_LM4.csv',index=False)
```
