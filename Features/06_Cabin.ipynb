```python
train_and_test = [train, test]
```

```python
train.isnull().sum()
```

```python
train['First Name']=train.Name.str.extract('([a-zA-Z]+)\,')
train.head()
```

```python
train['First Name'].value_counts()
```

```python
train.loc[:,['First Name']]
```

```python
for dataset in train_and_test:
    dataset['Cabin_Grade'] = dataset.Cabin.str.extract('([A-Z])')
    dataset.head()
    
```

```python
train['Cabin_Grade'].value_counts()
```

```python
for dataset in train_and_test:
    dataset['Cabin_Grade'] = dataset['Cabin_Grade'].fillna('S')
```

```python
pd.crosstab(train['Cabin_Grade'],train['Survived'])
```

```python
for dataset in train_and_test:
    dataset['Cabin_Score'] = dataset['Cabin_Grade'].replace(['A','B','C','D','E','F','G','T'], 1)
    dataset['Cabin_Score'] = dataset['Cabin_Score'].replace('S', 0)
```
